# Perceptron 

## AND operation

```
def step(x):
  if x >= 0:
    return 1
  else :
    return 0
```

```
def and_step(first, second):
  w1 = 1
  w2 = 1
  bias = -1.5
  result = w1 * first + w2 * second + bias
  return step(result)
```

```
print(1, 'and', 1, '=', and_step(1,1))
print(1, 'and', 0, '=', and_step(1,0))
print(0, 'and', 1, '=', and_step(0,1))
print(0, 'and', 0, '=', and_step(0,0))
```

## or operation

```
def or_step(first, second):
  w1 = 1
  w2 = 1
  bias = -0.5
  result = w1 * first + w2 * second + bias
  return step(result)
```

```
print(1, 'or', 1, '=', or_step(1,1))
print(1, 'or', 0, '=', or_step(1,0))
print(0, 'or', 1, '=', or_step(0,1))
print(0, 'or', 0, '=', or_step(0,0))
```

## NOT operation

```
def not_step(x):
  if x == 1:
    return 0
  else :
    return 1
```

```
print('not', 1, '=', not_step(1))
print('not', 0, '=', not_step(0))
```

## XOR operation

```
def xor_step(first, second):
  result = and_step(not_step(and_step(first, second)), or_step(first, second))
  return result
```

```
print(1, 'xor', 1, '=', xor_step(1,1))
print(1, 'xor', 0, '=', xor_step(1,0))
print(0, 'xor', 1, '=', xor_step(0,1))
print(0, 'xor', 0, '=', xor_step(0,0))
```

```
opertion_array = np.array([[1,1],[1,0],[0,1],[0,0]])
```

```
for first, second in opertion_array:
  print(first, 'xor', second, '=', xor_step(first,second))
```

## ReLU

```
def relu(x):
  x[x<0] = 0
  return x
```

```
arr2 = np.arange(-10,10,0.1)
```

```
y = relu(arr2)
```
