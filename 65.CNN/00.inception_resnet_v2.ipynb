```
from tensorflow.keras.applications.inception_resnet_v2 import InceptionResNetV2, decode_predictions

incep_resv2 = InceptionResNetV2(input_shape=(299,299,3))
# incep_resv2 = InceptionResNetV2(input_shape=(299,299,3),weights='imagenet', include_top=False, pooling='avg')
# incep_resv2.summary()
```

```
import tensorflow as tf
import time 

# image = tf.keras.preprocessing.image.load_img('./hand_8.jpg',  color_mode="grayscale", target_size=(28,28))			# with format information
image = tf.keras.preprocessing.image.load_img('./x_pred.jpg', target_size=(299,299))			# with format information
input_arr = tf.keras.preprocessing.image.img_to_array(image)	# return numpy

start = time.time() 
pred = incep_resv2.predict(input_arr.reshape(-1, 299, 299, 3))
time = time.time() - start

import numpy as np
label_key = np.argmax(pred)
label = decode_predictions(pred)
# label = label[0][0]

print("테스트 시 소요 시간 : {}".format(time))
# print('%s (%.2f%%)' % (label[1], label[2]*100))
# img = img[:,:,::-1]

# import matplotlib.pyplot as plt
# plt.figure(figsize=(11,11))
# plt.imshow(image)
# plt.axis("off")
# plt.show()
label_key, label
```

##load Dataset

```
import tensorflow as tf
_URL = 'https://storage.googleapis.com/mledu-datasets/cats_and_dogs_filtered.zip'

# path_to_zip = tf.keras.utils.get_file('cats_and_dogs.zip', origin=_URL, extract=True, )
path_to_zip = tf.keras.utils.get_file('cats_and_dogs.zip', origin=_URL, extract=True, cache_dir='./')
```

```
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(
        rescale=1./255,
        shear_range=0.6,
        zoom_range=0.6,
        # rotation_range=90,
        horizontal_flip=True,
        vertical_flip=True)
train_generator = train_datagen.flow_from_directory(
        './datasets/cats_and_dogs_filtered/train',
        # './testdatasets',
        target_size=(299, 299),
        batch_size=32,
        # class_mode='binary'
        )
```

```
x, y = train_generator.next()
```

```
x.shape, y.shape
```

##Fine Tuning

```
import tensorflow as tf
from tensorflow.keras import layers

# i = tf.keras.layers.Input([None, None, 3], dtype = tf.uint8)
i = tf.keras.layers.Input([None, None, 3])
# x = tf.cast(i, tf.float32)
# x = tf.keras.applications.inception_resnet_v2.preprocess_input(i)
x = tf.keras.applications.InceptionResNetV2()(i)      # InceptionResNetV2(include_top=False, pooling='avg')
# x = core(x)
outputs = layers.Dense(2, activation='softmax')(x)
# model = tf.keras.ModelModel(base_model.inputs, outputs)

model = tf.keras.Model(inputs=[i], outputs=[outputs])
```

```
# model.output
```

```
model.summary()
```

```
model.compile(optimizer='Nadam',
              loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```
model.fit(train_generator)
```

```
image = tf.keras.preprocessing.image.load_img('./dog01.jpg', target_size=(299,299))			# with format information
input_arr = tf.keras.preprocessing.image.img_to_array(image)	# return numpy

# start = time.time() 
pred = model.predict(input_arr.reshape(-1, 299, 299, 3))
# time = time.time() - start

import numpy as np
label_key = np.argmax(pred)
```
