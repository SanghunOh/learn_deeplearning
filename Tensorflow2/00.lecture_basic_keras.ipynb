##Tokenrizer

```
from tensorflow.keras.preprocessing.text import Tokenizer

text="댓글은 인터넷 게시물 밑에 남길 수 있는 짧은 글이다."
list_text=[["댓글은 인터넷 게시물 밑에 남길 수 있는 짧은 글이다."], ['선한 영향력 굿.']]

tokenizer = Tokenizer()

tokenizer.fit_on_texts([text])
# tokenizer.fit_on_texts(list_text)

tokenizer.word_index
```

```
sub_text="유즈넷(USENET) 뉴스그룹 초창기 시절 'reply'를 '리플라이', '답장' 등으로 주로 사용하였다"
encoded=tokenizer.texts_to_sequences([sub_text])[0]
print(encoded)
```

```
sub_text="보통 인터넷 게시물 밑에는 댓글란이 있어 그 게시물과 관련하여 독자는 의견을 표할 수 있다"
encoded = tokenizer.texts_to_sequences([sub_text])[0]
print(encoded)
```

##interger Encoding

```
from tensorflow.keras.preprocessing.text import Tokenizer
sentences=[['barber', 'person'], ['barber', 'good', 'person'], ['barber', 'huge', 'person'], ['knew', 'secret'], ['secret', 'kept', 'huge', 'secret'], ['huge', 'secret'], ['barber', 'kept', 'word'], ['barber', 'kept', 'word'], ['barber', 'kept', 'secret'], ['keeping', 'keeping', 'huge', 'secret', 'driving', 'barber', 'crazy'], ['barber', 'went', 'huge', 'mountain']]
tokenizer = Tokenizer(num_words = 5 + 1, oov_token='OOV')
# 빈도수 상위 5개 단어만 사용. 숫자 0과 OOV를 고려해서 단어 집합의 크기는 +2
tokenizer.fit_on_texts(sentences)
print(tokenizer.word_index)
```

##Embadding

```
import tensorflow as tf
import numpy as np

model = tf.keras.Sequential()
model.add(tf.keras.layers.Embedding(input_dim=1000, output_dim=1)) #,input_length=10))
# dimension.  
input_array = np.random.randint(1000, size=(32, 10))
print(input_array.shape, input_array, )
model.compile('rmsprop', 'mse')
output_array = model.predict(input_array)
print(output_array.shape, output_array)
```
