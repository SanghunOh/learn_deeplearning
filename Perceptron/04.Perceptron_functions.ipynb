## AND

```
# AND 퍼셉트론
# 1 1 -> 1
# 1 0 -> 0
# 0 1 -> 0
# 0 0 -> 0
def step(output):
  if output < 0:
    return 0
  else:
    return 1

# unfix AND
def AND(a, b):
  w1 = 0
  w2 = 0
  bias = 0 # y절편
  output = a * w1 + b * w2 + bias
  return step(output)

print(AND(1, 1))
print(AND(1, 0))
print(AND(0, 1))
print(AND(0, 0))
```

```
def AND_fix(a, b):
  w1 = 0.5
  w2 = 0.5
  bias = -0.6
  output = a * w1 + b * w2 + bias
  # print(w1,w2,bias,output)
  return step(output)

print(AND_fix(1, 1))
print(AND_fix(1, 0))
print(AND_fix(0, 1))
print(AND_fix(0, 0))
```

##OR

```
def OR_fix(a, b):
  w1 = 0.5
  w2 = 0.5
  bias = -0.5
  output = a * w1 + b * w2 + bias
  # print(w1,w2,bias,output)
  return step(output)

print(OR_fix(1, 1))
print(OR_fix(1, 0))
print(OR_fix(0, 1))
print(OR_fix(0, 0))
```

##NOT

```
def NOT(x):
  if x == 1:
    return 0
  else:
    return 1

print(NOT(1))
print(NOT(0))    
```

##XOR

```
def XOR(a,b):
  return AND_fix(NOT(AND_fix(a,b)),OR_fix(a, b))

print(XOR(1, 1))
print(XOR(1, 0))
print(XOR(0, 1))
print(XOR(0, 0))  
  
```

#Sigmoid function


```
import matplotlib.pyplot as plt

def sigmoid(x):
  return 1/(1+np.exp(-x))

x = np.arange(-10, 10, 0.1)
plt.plot(x, sigmoid(x))
```
