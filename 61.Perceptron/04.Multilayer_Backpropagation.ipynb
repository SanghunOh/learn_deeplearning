# Backpropagation

```
import numpy as np
import matplotlib.pyplot as plt
```

```
def sigmoid(x):
  return 1 / (1+np.exp(-x))

def sigmoid_prime(x):
  return x * (1-x)
```

```
input1 = np.array([np.arange(4), np.arange(4)])
input1
```

```
# 4 x 3 x 2
n_records, n_inputs = input1.shape
n_records, n_inputs
```

```
n_hidden = 3
weights_input_to_hidden = np.random.normal(0, n_inputs**-5, size=(n_inputs, n_hidden))
weights_input_to_hidden
```

```
n_output = 2
weights_input_to_output = np.random.normal(0, n_inputs**-5, size=(n_hidden, n_output))
weights_input_to_output
```

```
hidden1 = sigmoid(np.dot(input1, weights_input_to_hidden))
hidden1_output = sigmoid(np.dot(hidden1, weights_input_to_output))
hidden1_output
```

# backpropagation

```
input2 = np.array([0.1, 0.3])
```

```
predict = 1
```

```
weights = np.array([0.4, -0.2])
```

```
hidden_weight = 0.1
```

```
hidden_output = sigmoid(np.dot(input2, weights))
```

```
output = sigmoid(np.dot(hidden_output, hidden_weight))
output
```

```
error = predict - output
```

```
err_term = error * sigmoid_prime(output)
err_term
```

```
hidden_err_term = hidden_weight * err_term * sigmoid_prime(output)
hidden_err_term
```

```
learning_rate = 0.5
```

```
delta_w = learning_rate * err_term * hidden_output
delta_w_hidden = learning_rate * hidden_err_term * input2
delta_w, delta_w_hidden
```
