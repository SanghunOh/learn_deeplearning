## Test CNN with cifar10

```
# from __future__ import print_function
import tensorflow as tf
```

```
from tensorflow.keras.datasets import cifar10
# from tensorflow.keras.preprocessing.image import ImageDataGenerator # overfit 
from tensorflow.keras import layers
import os
```

```
save_dir = os.path.join(os.getcwd(), 'saved_models')
model_name = 'keras_cifar10_trained_model.h5'
```

```
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
print(x_train.shape, 'x_train shape')
print(x_train.shape[0], 'x_train sample')
print(x_test.shape[0], 'x_test sample')

import matplotlib.pyplot as plt
plt.imshow(x_train[1])

type(x_train.shape), y_train.shape
```

```
num_classes = 10 # 특징 갯수

y_train = tf.keras.utils.to_categorical(y_train, num_classes) # one hot encoding
y_test = tf.keras.utils.to_categorical(y_test, num_classes) # one hot encoding

print('x_train', y_train[0])
# print('y_train', y_test[0])
```

```
# 네트워크 설계

model = tf.keras.models.Sequential()
model.add(layers.Conv2D(32, (3,3), padding='same',
                input_shape=x_train.shape[1:])) # 32, 32, 3
model.add(layers.Activation('relu'))

model.add(layers.Conv2D(32, (3,3)))
model.add(layers.Activation('relu'))
          
model.add(layers.MaxPooling2D(pool_size=(2,2)))
model.add(layers.Dropout(0.25))
model.add(layers.Flatten()) # 데이터를 펴는 작업

model.add(layers.Dense(512))
model.add(layers.Activation('relu'))
model.add(layers.Dropout(0.5))

model.add(layers.Dense(num_classes))
model.add(layers.Activation('softmax'))
model.summary()
```

```
batch_size = 164 # 입력 묶음 사이즈
epochs = 2 # 역전파 횟수
data_augmentation = True
num_predictions = 20

# opt = tf.keras.optimizers.RMSprop(lr=0.0001)
# opt = tf.keras.optimizers.Adam(lr=0.0001)
opt = tf.keras.optimizers.Nadam(lr=0.0001)

model.compile(loss='categorical_crossentropy',
             optimizer=opt,
             metrics=['accuracy'])

model.fit(x_train, y_train,
         batch_size=batch_size,
         epochs=epochs,
         verbose=1,
         validation_data=(x_test, y_test))
# epochs = 50
# RMSprop - 2s 8ms/step - loss: 0.5062 - accuracy: 0.8322 - val_loss: 1.0074 - val_accuracy: 0.6968
# Adam   - 6s 20ms/step - loss: 0.2063 - accuracy: 0.9284 - val_loss: 1.0622 - val_accuracy: 0.7021
# Nadam  - 8s 25ms/step - loss: 0.1716 - accuracy: 0.9407 - val_loss: 1.0705 - val_accuracy: 0.7069
```

```
prob = model.predict(x_train)
prob[0]
```

```
model.evaluate(x_train, y_train, verbose=2)
```
